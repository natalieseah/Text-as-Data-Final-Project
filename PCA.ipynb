```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.decomposition import PCA
import plotly.express as px
```

```python
TFIDF_L2 = pd.read_csv("TFIDF_L2.csv", sep="|", index_col=0)
LIB = pd.read_csv("LIB.csv", sep="|", index_col=0)
```

# PCA Components (4)

```python
pca_engine = PCA(n_components=10)

X_pca = pca_engine.fit_transform(TFIDF_L2)

COMPONENTS = pd.DataFrame(
    pca_engine.components_,
    columns=TFIDF_L2.columns,
    index=[f'PC{i}' for i in range(pca_engine.n_components_)]
)
COMPONENTS
```

```python
COMPONENTS.loc['PC1'].sort_values(ascending=False).head(5)
```

```python
COMPONENTS.loc['PC2'].sort_values().head(5)
```

```python
COMPONENTS.to_csv("/Users/natalieseah/Desktop/TextasData/PCA_COMPONENTS.csv", sep="|")
```

# PCA DCM (4)

```python
DCM = pd.DataFrame(
    X_pca,
    index=TFIDF_L2.index,
    columns=[f'PC{i}' for i in range(X_pca.shape[1])]
)
DCM
```

```python
DCM.to_csv("/Users/natalieseah/Desktop/TextasData/PCA_DCM.csv", sep="|")
```

# PCA Loadings (4)

```python
LOADINGS = pd.DataFrame(
    pca_engine.components_.T * np.sqrt(pca_engine.explained_variance_),
    index=TFIDF_L2.columns,
    columns=[f'PC{i}' for i in range(pca_engine.n_components_)]
)

LOADINGS.index.name = 'term_str'

LOADINGS
```

```python
LOADINGS.to_csv("/Users/natalieseah/Desktop/TextasData/PCA_LOADINGS.csv", sep="|")
```

# PCA Visualization 1

```python
# Join with LIB to get artist labels
DCM_LABELED = DCM.join(LIB[['Artist', 'Title']])

# Documents scatterplot
fig = px.scatter(
    DCM_LABELED,
    x='PC0',
    y='PC1',
    color='Artist',
    hover_name='Title',
    hover_data={'Artist': True, 'PC0': False, 'PC1': False},
    title='Documents in PC0 x PC1 Space (colored by Artist)',
    height=800,
    opacity=0.6
)
fig.show()

# Loadings scatterplot
COMPS_RESET = COMPS.reset_index()

fig2 = px.scatter(
    COMPS_RESET,
    x='PC0',
    y='PC1',
    hover_name='term_str',
    title='Loadings in PC0 x PC1 Space',
    height=800,
    opacity=0.4
)
fig2.show()
```
